```python
import plotly.express as px

gapminder = px.data.gapminder()
gapminder.head(5)

```

```python
fig = px.choropleth(gapminder,
                    locations ="iso_alpha",
                    color ="lifeExp",
                    hover_name ="country", 
                    color_continuous_scale = px.colors.sequential.Plasma,
                    scope ="world",
                    animation_frame ="year")
fig.show()
```

```python
fig = px.bar(gapminder, 
             x ="continent", 
             y ="pop",
             color ='lifeExp',
             animation_frame ='year',
             hover_name ='country', 
             range_y =[0, 4000000000])
fig.show()
```

```python
import plotly.express as px

df = px.data.gapminder()

fig = px.scatter(df,
                 x="gdpPercap",
                 y="lifeExp",
                 size="pop",
                 color="continent",
                 hover_name="country",
                 log_x=True,
                 size_max=60,
                 animation_frame="year",
                 animation_group="country")

fig.show()
```

### Removing three column New Cases, New deaths, New Recovered

```python
dataset1.drop(['New cases','New deaths', 'New recovered'],axis = 1,inplace = True)  ## inplace true dene s  permanently delete hoga
```

### sort the column

```python
dataset2 = dataset1.sort_values(by="Confirmed", ascending=False)

dataset2

```

### Creating table using plotly express

```python
# Import create_table Figure Factory

from plotly.figure_factory import create_table

# colorscale = [[0, '#4d004c'], [.5, '#f2e5ff'], [1, '#ffffff']]
color_continuous_scale=[[0, '#4d004c'], [0.2, '#9932cc'], [0.6, '#f2e5ff'], [1, '#ffffff']]

table = create_table(dataset2.head(15), colorscale=color_continuous_scale)
py.iplot(table)
```

### Bar graph:- Country/region vs Confirmed case

```python
px.bar(dataset2.head(15), x = 'Country/Region', 
       y = 'Confirmed',color = 'Confirmed', 
       height = 500,hover_data = ['Country/Region', 'WHO Region'])
```

### Here Alignment is Opposite

```python
px.bar(dataset2.head(15), x = 'Confirmed', 
       y = 'Country/Region',color = 'Confirmed', orientation= 'h',
       height = 500,hover_data = ['Country/Region', 'WHO Region'])
```
